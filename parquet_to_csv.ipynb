```python
# Путь к папке с файлами parquet
folder_path = 'C:\\Users\\VadimTyr\\otus_dwh_project\\data\\raw\\yellow\\'

# Собираем все файлы parquet из папки в один DataFrame
dfs = []
for file in os.listdir(folder_path):
    if file.endswith('.parquet'):
        file_path = os.path.join(folder_path, file)
        df = pd.read_parquet(file_path)
        dfs.append(df)
# Путь для сохранения CSV файла
csv_save_path = r'C:\Users\VadimTyr\otus_dwh_project\seeds\yellow_taxi_tripdata.csv'
combined_df = pd.concat(dfs, ignore_index=True)
# Преобразуем объединенный DataFrame в CSV и сохраняем его по указанному пути
combined_df.to_csv(csv_save_path, index=False)
```
